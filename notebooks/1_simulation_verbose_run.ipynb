# Simulation 1 — Single Verbose Run

This simulation runs a single 6-player game with verbose logging enabled. The purpose is twofold: to verify that the engine runs correctly end-to-end, and to provide a readable trace of one complete game that illustrates how the round loop, agent decisions, and card draws interact in practice.

A fixed random seed is used so the output is stable and reproducible.

```python
random.seed(0)
record = engine.run_game(player_count = 6, verbose = True)
```

```python
print("── Game Record ─────────────────────────────────────")
print(f"Outcome:                  {record.outcome.upper()}")
print(f"Players:                  {record.player_count}")
print(f"Characters:               {[c.value for c in record.characters]}")
print(f"Rounds played:            {record.rounds_played}")
print(f"Boat parts built:         {record.boat_parts_built} / {record.boat_parts_required}")
print(f"Volcano cards remaining:  {record.volcano_cards_remaining}")
print(f"Final scores:             { {c.value: s for c, s in record.final_scores.items()} }")
```

### Reading the Verbose Trace

Each row in the trace above represents one round:

* **`Round N`** - round number
* **`mission: <name>`** - the mission completed this round; `—` means the round ended without a completion (mission failed or no eligible mission due to Panic)
* **`volcano left: N (used M)`** - cards remaining in the volcano deck and how many were consumed this round (0 on a success, 1 on a failure or special card effect)
* **`boat: X/Y`** - boat parts built so far vs. the number required to win
* **`scores: [...] (+[...])`** - each player's cumulative score and the points gained this round (negative values indicate a Smoke penalty was applied)
