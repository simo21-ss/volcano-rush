## Agent-Based Modeling & Monte Carlo Simulations

**Source:** https://studyguides.com/study-methods/overview/clzggcm6j0j8t8cfe9zugo2t8

For instance, in Agent-Based Modeling (ABM), each agent operates according to a set of predefined rules, interacting with other agents and the environment. These interactions can lead to complex emergent phenomena that provide insights into the collective behavior of the system. Monte Carlo simulations, on the other hand, use random sampling to explore the range of possible outcomes in a given scenario, offering probabilistic insights into the impact of uncertainty and risk on strategic decisions.

Another controversy surrounds the assumptions of rationality in game theory. Traditional game theory assumes that agents are rational and will always act to maximize their utility. However, in real-world scenarios, human behavior is often influenced by irrational factors, emotions, and imperfect information. This limitation has led to criticisms of game theory simulations, as they may not fully capture the complexities of human decision-making.

## Research Gaming & ABM Integration

**Source:** https://www.sciencedirect.com/science/article/pii/S0198971524000711

Research gaming involves "human beings acting as themselves or playing simulated roles in an environment which is either actual or simulated" that treats games as models in which the decisions, actions and behaviours of players inform understanding of real-world systems (Shubik, 1975 p. 13).

Researchers have also integrated strategic gaming experiments with ABMs. Deadman, Schlager, and Gimblett (2000) used participant exit interviews from a common-pool resource experiment to design a list of strategic options that are used by agents in an ABM.

When two equally good options are available to an agent (i.e., both will get the agent equally close to their destination), it makes a probabilistic decision about which intersection it prefers based on the type (light, zebra stripe, sign, no control) at each step in the simulation. The probabilities are based on the frequency data from the players playing the game.

## Game Balance — Four Definitions

**Source:** https://gamedesignconcepts.wordpress.com/2009/08/20/level-16-game-balance/

In my experience, when we talk of "game balance" we are generally talking about one of four things. Context usually makes it clear which one we are talking about:

- In single-player games, we use "balance" to describe whether the challenge level is appropriate to the audience;
- In multi-player games where there is asymmetry (that is, where players do not start with exactly equal positions and resources), we use "balance" to describe whether one starting position is easier to win with than another.
- Within a game, if there are multiple strategies or paths to victory that can be followed within the game, we use "balance" to describe whether following one strategy is better or worse than following another.
- Within a system that has several similar game objects (such as cards in a trading-card game, weapons in a role-playing game, and so on), we use "balance" to describe the objects themselves, specifically whether different objects have the same cost/benefit ratio.

Sometimes, players are so different that direct comparisons are impossible. Some games not only give players different starting resources and positions, but also different rules to play by. Some players may have exclusive access to certain resources or abilities. One common asymmetry in games is to give players different and conflicting objectives (for example, one team's objective is to survive for some number of turns and the other team's objective is to eliminate the first team before that many turns). The more difficult it is to make direct comparisons between players, the more you will have to playtest to compensate.

The problem here is that, once a dominant strategy is discovered, astute players will ignore all suboptimal strategies. Everything in the game that is not part of the dominant strategy becomes extraneous noise. There is nothing inherently wrong with a game that has a single winning strategy, but in this case the suboptimal ones should be removed to make the game more streamlined. If you include options for players that are suboptimal, these become false decisions, because really there is only one decision (follow the dominant strategy). If it is worth including several potential winning strategies in a game, then, it becomes much more interesting if those strategies are balanced.

In all of these cases, there are two goals. The first is to prevent any game object from being so weak that it is useless in comparison with other objects. The second goal is to prevent a game object from being too powerful. Any single game object that becomes a dominant strategy makes all other objects in the game useless in comparison. Two objects are balanced if they have the same cost/benefit ratio.

## Board Game Balance Metrics

Essential board game balance metrics include win rates per strategy/faction (aiming for 45–55%), average score differentials, first-player advantage, and the standard deviation of final scores. Crucial qualitative data includes tracking "feels unfair" moments, strategy diversity, and identifying runaway leaders.

### Key Quantitative Metrics

- **Win Rates:** Track the win percentage of each character, faction, or strategy. A balanced game usually sees all options winning roughly equally, ideally within 45–55% for highly competitive games.
- **Average Score Differential:** Measure how close the final scores are. Too high a difference indicates a "runaway leader" problem; too low might mean choices don't matter.
- **First Player Advantage:** Measure if the person going first wins significantly more often.
- **Standard Deviation of Scores:** High deviation suggests high luck or extreme strategy power differences.
- **Resource/Action Efficiency:** Evaluate the average value of a turn or action (e.g., Victory Points per gold spent).
- **Strategy Diversity:** Count how many different paths to victory are viable. If 90% of winners use one strategy, the game is unbalanced.

### Key Qualitative Metrics (Playtesting)

- **"Feels Unfair" vs. "Actually Unbalanced":** Distinguish between a lucky moment and a consistently broken strategy.
- **Player Agency:** Did losers feel they made bad choices, or that the game gave them no chance?
- **Time to Win/Kill:** How long does it take for a strategy to become dominant?

### Balancing Approaches

- **Relative Values:** Set a baseline (e.g., cheapest card) and ensure more powerful items cost proportionally more.
- **Negative Feedback Loops:** Design mechanics where players who fall behind get bonuses, or leaders face disadvantages (e.g., "rubber-banding").
- **Asymmetry Management:** In asymmetric games, balance is achieved through unique strengths and weaknesses that force players into "relational balance," where they counter each other.
